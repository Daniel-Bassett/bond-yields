```python
import pandas as pd
import numpy as np 
import json

import requests
from bs4 import BeautifulSoup
from lxml import html

import plotly_express as px
```

```python
url = 'https://home.treasury.gov/resource-center/data-chart-center/interest-rates/TextView?type=daily_treasury_yield_curve&field_tdr_date_value=2023'
```

```python
response  = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
tree = html.fromstring(response.content)
```

```python
dates = soup.find_all('td', class_=)
dates = [date.text for date in dates]
```

```python
def create_col(col_name, class_name):
    df = soup.find_all('td', class_=class_name)
    df = [{col_name: price.text} for price in df]
    df = pd.DataFrame(df)
    return df
```

```python
date = create_col('date', 'views-field views-field-field-tdr-date')
one_month = create_col('one_month', 'bc1month views-field views-field-field-bc-1month')
two_month = create_col('two_month', 'bc2month views-field views-field-field-bc-2month')
three_month = create_col('three_month', 'bc3month views-field views-field-field-bc-3month')
four_month = create_col('four_month', 'bc4month views-field views-field-field-bc-4month')
six_month = create_col('six_month', 'views-field views-field-field-bc-6month')
```

```python
prices = pd.concat([date, one_month, two_month, three_month, four_month, six_month], axis=1)
prices = (prices
          .assign(date=lambda df: pd.to_datetime(df['date'].str.strip()))
          .astype({'one_month': float, 'two_month': float,
                   'three_month': float, 'four_month': float,
                   'six_month': float
                   })
          )
```

```python
px.line(prices.pct_change().dropna())
```
